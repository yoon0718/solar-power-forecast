```python
from datetime import datetime
```

```python
data = test_data_Gangwon.copy()
```

```python
data.set_index('TM', inplace=True)
```

```python
data
```

```python
# ARIMA 모델 학습 및 예측
predictions = []  # 예측 결과를 저장할 리스트

# ARIMA 모델 파라미터 설정
p, d, q = 2, 2, 0  # ARIMA 모델의 차수 설정
forecast_period = 365  # 향후 1년을 예측하기 위한 기간 설정

for _ in range(forecast_period):
    # ARIMA 모델 학습
    model = ARIMA(data['SI'], order=(p, d, q))
    fit_model = model.fit()

    # 다음날의 날씨 예측
    forecast = fit_model.forecast(steps=1)
    next_day_forecast = forecast[0]

    # 예측 결과를 리스트에 추가
    predictions.append(next_day_forecast)

    # 이전 예측 결과를 데이터에 추가하여 다음 예측에 사용
    data = pd.concat([data, pd.DataFrame([next_day_forecast], columns=['SI'], index=[pd.Timestamp(data.index[-1]) + pd.DateOffset(days=1)])])

# 예측 결과를 데이터프레임으로 변환
forecast_index = pd.date_range(start=data.index[-1], periods=forecast_period, freq='D')
forecast_df = pd.DataFrame(data=predictions, index=forecast_index, columns=['SI'])

# 원래 데이터와 예측 결과를 합치기
combined_data = pd.concat([data, forecast_df])
```
