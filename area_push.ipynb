```python
def process_and_extract_columns(input_df):
    # 'TM' 컬럼에서 'year', 'month', 'day', 'hour'를 추출하여 새로운 컬럼에 저장
    input_df['year'] = 0
    input_df['month'] = 0
    input_df['day'] = 0
    input_df['hour'] = 0
    for i in range(len(input_df)):
        input_df['year'].iloc[i] = str(input_df['TM'].iloc[i])[:4]
        input_df['month'].iloc[i] = str(input_df['TM'].iloc[i])[4:6]
        input_df['day'].iloc[i] = str(input_df['TM'].iloc[i])[6:8]
        input_df['hour'].iloc[i] = str(input_df['TM'].iloc[i])[8:10]

    # 'TM'과 'SI'만 포함하는 새로운 테이블 생성
    new_table = input_df[['TM','year','month','day','hour','SI']]
    
    return new_table
```

```python
interval = 3
def make_data(data):
    x = []
    y = []
    temps = list(data["SI"])
    for i in range(len(temps)):
        if i <= interval: continue
        y.append(temps[i]) #4부터 리스트에 추가
        xa = []
        for p in range(interval):
            d = i + p - interval
            xa.append(temps[d])
        x.append(xa)
    return (x, y)

x, x1 = make_data(my_area_data_wdm_daejeon_SI_df)
```
