```python
import requests
import pandas as pd
import json
import re

apikey = '' #인증키
data_list = []

for year in range(2020, 2024):
    for month in range(1, 13):
        formatted_year = str(year)
        formatted_month = f"{month:02d}"

        tm1 = f"{formatted_year}{formatted_month}010000"
        tm2 = f"{formatted_year}{formatted_month}312300"

        url = f'https://apihub.kma.go.kr/api/typ01/url/kma_sfctm3.php?tm1={tm1}&tm2={tm2}&stn=0&help=0&authKey={apikey}'

        json_response = requests.get(url)
        a = json_response.text.split("\n")
        a = a[2:]
        a = a[:1] + a[2:]
        data_lines = a[1:]

        columns = re.split(r'\s+', a[0].strip())[1:]

        for line in data_lines:
            values = re.split(r'\s+', line.strip())
            data_list.append(values)

df = pd.DataFrame(data_list, columns=columns)

df
```
