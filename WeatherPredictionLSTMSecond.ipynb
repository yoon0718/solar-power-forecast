```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

```python
data.set_index('TM', inplace=True)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data['SI'], label='일사량')
plt.title('일사량 변화')
plt.xlabel('날짜')
plt.ylabel('일사량')
plt.legend()
plt.show()
```

```python
train_size = int(len(data) * 0.7)
train_data = data['SI'][:train_size]
test_data = data['SI'][train_size:]


# LSTM 모델을 통한 기상 예측
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(np.array(data['SI']).reshape(-1, 1))

def create_dataset(dataset, time_steps=1):
    X, Y = [], []
    for i in range(len(dataset) - time_steps - 1):
        a = dataset[i:(i + time_steps), 0]
        X.append(a)
        Y.append(dataset[i + time_steps, 0])
    return X, Y

time_steps = 7
X, Y = create_dataset(scaled_data, time_steps)
#X = np.reshape(X, (np.array(X).shape[0], np.array(X).shape[1], 1))
```

```python
X = pd.DataFrame(X)
```

```python
X
```

```python
X_train = []
y_train = []
for i in range(int(len(X)*0.75)-7):
    X_train.append(X.iloc[i:i+7,:])
    y_train.append(X.iloc[i+7,:])
X_train = np.array(X_train)
y_train = np.array(y_train)

X_val = []
y_val = []
for i in range(int(len(X)*0.75)-7,len(X)-7):
    X_val.append(X.iloc[i:i+7,:])
    y_val.append(X.iloc[i+7,:])
X_val = np.array(X_val)
y_val = np.array(y_val)
```

```python
X_train.shape, y_train.shape
```

```python
# LSTM 모델 정의
model = Sequential()
model.add(LSTM(64, input_shape=(7, 7), return_sequences=True))
model.add(LSTM(32))
model.add(Dense(7))
model.compile(optimizer='adam', loss='mse', metrics=['accuracy', 'mse'])
```

```python
# 모델 훈련
model.fit(X_train, y_train, validation_data=(X_val,y_val), epochs=100, batch_size=32, verbose=2)
```

```python
X_val.shape
```

```python
# 향후 1년간의 데이터 예측
future_predictions = []
current_batch = X_val[-1].reshape((1, 7, 7)).copy()
for i in range(365):  # 1년간의 데이터 예측
    future_prediction = model.predict(current_batch)[0]
    future_predictions.append(future_prediction)
    current_batch = np.append(current_batch[:, 1:, :], [[future_prediction]], axis=1)

```

```python

# 시각화
plt.figure(figsize=(12, 6))
#plt.plot(data.index[-len(y_val):], y_val, label='실제 데이터')
plt.plot(pd.date_range(start=data.index[-1], periods=len(future_predictions), freq='D'), future_predictions, label='예측 데이터')
plt.title('기상 데이터 예측')
plt.xlabel('날짜')
plt.ylabel('종합 예측')
plt.legend()
plt.show()
```
